```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline   
```

```python
data = pd.read_csv("/Users/Owner/Desktop/FanGraphs_Alex_Cobb_Velo_2017.csv")
```

```python
data.columns
```

```python
data = data.sort_index(ascending=False)
data.tail()
```

```python
data.drop([0], axis=0, inplace=True)
```

```python
data.plot(x='Date', y='FBv', figsize=(15, 5))
```

```python
data.plot(x="Date", y=['FBv', 'CBv', 'CHv'], figsize=(15, 5))
plt.xlabel('Date')
plt.ylabel('Velocity')
plt.title('Alex Cobb Season Velocity')
```

```python
#Remove percent sign 

data['FB%'] = data['FB%'].replace({'\%':''}, regex = True)
data['CB%'] = data['CB%'].replace({'\%':''}, regex = True)
data['CH%'] = data['CH%'].replace({'\%':''}, regex = True)

data.rename(columns={'FB%':'FB_p', 'CB%':'CB_p', 'CH%':'CH_p'}, inplace=True)
```

```python
data['FB_p'] = data['FB_p'].astype(float)
data['CB_p'] = data['CB_p'].astype(float)
data['CH_p'] = data['CH_p'].astype(float)
```

```python
data.plot(x="Date", y=['FB_p', 'CB_p', 'CH_p'], figsize=(15, 5))
plt.xlabel('Date')
plt.ylabel('Pitch Type (%)')
plt.title('Alex Cobb Season Pitch Usage')
```
